### Prediction

```python
clean_df = df.dropna()
```

```python
clean_df.shape #Because of large numbers of missing values in the predicted variable, NaNs were dropped
```

```python
clean_df
```

```python
keep_list = ['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Critic_Score']
clean2 = clean_df[keep_list]
clean2.head()
```

```python
import seaborn as sns; sns.set()

sns.pairplot(clean2, hue = 'Critic_Score', size = 3)
#g = sns.pairplot(clean2, hue = 'Critic_Score', size = 3, kind = 'reg')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.cross_validation import train_test_split
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = LinearRegression(fit_intercept = True)
```

```python
clean2.head()
```

```python
clean2_x = clean2.drop('Critic_Score', axis = 1)
clean2_y = clean2['Critic_Score']
print(clean2_x.shape, clean2_y.shape)
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(clean2_x, clean2_y, random_state=1)
```

```python
Xtrain.head()
```

```python
Xtest.head()
```

```python
ytrain[:5]
```

```python
ytest[:5]
```

```python
model.fit(Xtrain, ytrain)
```

```python
model.coef_ #coefficient of each variable
```

```python
model.intercept_ #intercept of slope
```

```python
y_model = model.predict(Xtest)
```

```python
import math
mse = mean_squared_error(ytest, y_model)
rmse = math.sqrt(mse)
```

```python
print("Root mean squared error: %.2f"
      % rmse)
```

```python
print('Variance score: %.2f' % r2_score(ytest, y_model))
```
